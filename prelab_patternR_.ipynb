```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print(df.head())
```

```python
print(df.shape)
print(df.describe())
print(df.info())
```

```python
import matplotlib.pyplot as plt

df.hist(figsize=(10,8))
plt.show()
```

```python
plt.scatter(
    df['sepal length (cm)'],
    df['petal length (cm)']
)

plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Iris Dataset Scatter Plot")
plt.show()
```
